```python
import pandas as pd

df = pd.read_csv("../data/raw/synthetic_crudop_logs.csv")
```

```python
import numpy as np

df["timestamp"] = pd.to_datetime(df["timestamp"])
df["hour"] = df["timestamp"].dt.hour

# cyclical encoding (VERY IMPORTANT)
df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
```

```python
df = pd.get_dummies(df, columns=["operation"])
```

```python
df.head()
```

```python
df["total_vol"] = df["success_vol"] + df["fail_vol"]

df["fail_ratio"] = df["fail_vol"] / (df["total_vol"] + 1)

df["latency_gap"] = df["success_rt_avg"] - df["fail_rt_avg"]

df["vol_per_latency"] = df["total_vol"] / (df["success_rt_avg"] + 1)
```

```python
df.head()
```

```python
df = df.dropna()
df = df.drop(columns=["timestamp"])
df.head()
```

```python
df.to_csv("../data/processed/features_crudop_logs.csv", index=False)
```
